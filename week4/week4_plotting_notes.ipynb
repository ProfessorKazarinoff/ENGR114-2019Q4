### Build a line plot with a couple customizations

 * title
 * x-axis label
 * y-axis label
 * red line

```python
# Create data to plot
x = np.arange(0,np.pi*4,0.1)
y = np.sin(x)

# Create the fig, ax objects
fig, ax = plt.subplots()

# Plot the data and customize the plot
ax.plot(x,y,'r')
ax.set_title('Line plot of the sine function')
ax.set_xlabel('angles x from 0 to 4pi')
ax.set_ylabel('Sin(x)')

# Show the plot
plt.show()
```

### Create a multi-line plot and save it
 
  * sine and cosine functions, sin a red line, cosine, blue line
  * x-axis label
  * y-axis label
  * title
  * legend
  * save the plot 

```python
# create the x array, and two y arrays for sine and cosine
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# create fig, ax objects
fig, ax = plt.subplots()

# plot the two y arrays using the same x-array on the same ax object
ax.plot(x,y1,'r')
ax.plot(x,y2,'b')

# customize the plot
ax.set_xlabel('x values 0 to 4$\pi$')
ax.set_ylabel('sine and cosine of x')
ax.set_title('Sine and Cosine of x')
ax.legend(['sin','cos'], loc='upper right')  # include a list of strings

# save the plot
plt.savefig('multi-line_plot.png', dpi=300, bbox_inches='tight')  #300 dpi for word or pdf's, 72 dpi for websites

# show the plot
plt.show()
```

### Build a plot with 3 lines and axis limits

```python
# data to plot
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# fig, ax objects
fig, ax = plt.subplots()

# plot all three y-values and customize
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
ax.set_xlim([-1,13])
ax.set_ylim([-2,2])
ax.legend(['sine','cosine','tangent'])

# save and show the plot
plt.savefig('3_line_plot.png',dpi=72,bbox_inches='tight')

plt.show()
```

### Bar Charts and Pie Charts

#### Build a pie chart

 * 8 MacOS users
 * 12 Windows users
 * 1 Linux user

```python
# construct lists of the data
labels = ['MacOS','Windows10','Linux']
sizes = [8,12,1]
explode = (0.05,0.3,0.05)
# create fig and ax objects
fig, ax = plt.subplots()

# ax.pie to create the plot and customize
ax.pie(sizes, labels=labels, autopct='%1.0f%%', explode=explode)
ax.axis('equal')
ax.set_title('OS used by ENGR114 students')

# show plot
plt.show()
```

#### Create a bar chart

 * 3 Cummutes less than 20 min
 * 6 Commutes between 20 and 40 min
 * 12 Commutes greater than 40 min

```python
# bar hieghts array, bar location arrays, bar label arrays
heights = np.array([3,6,12])
bar_locs = [0,1,2]
bar_labels = ['less than 20','20 to 40','more than 40']

# create fig, ax objects
fig, ax = plt.subplots()

# ax.bar method to build the plot and customize
ax.bar(bar_locs,heights, align='center')
ax.set_title('Number of students who have different communte times')
ax.set_ylabel('Number of Students')
ax.set_xlabel('Commute Time Range (minutes)')
ax.set_xticks(bar_locs)
ax.set_xticklabels(bar_labels)

# save the plot
plt.savefig('bar_plot.png')

# show the plot
plt.show()
```

#### Build a histogram plot
 
 * psudo-random numbers with a gaussian distribution
 * label the x-axis with 'bin range'
 * label the y-axis with 'frequency'
 * title the plot 'A Histogram'


```python
# create an array with psudo-random numbers that have a gaussian distribution
mu = 20
sigma = 5
x = np.random.normal(mu, sigma, size=200)

# create the figure and axis objects
fig, ax = plt.subplots()

# create the histogram and customize
ax.hist(x, 10)       # x is the array of values, 10 is the number of bins
ax.set_xlabel('bin range')
ax.set_ylabel('frequency')
ax.set_title('A histogram')

# show plot
plt.show()
```

#### Build a scatter plot

 * two arrays of semi-focused random numbers
 * title as 'A scatter plot of two semi-focused arrays of random numbers'

```python
# create the arrays of semi-focused random numbers
x1 = 2*np.random.randn(300) + 15
y1 = 2*np.random.randn(300) + 12

x2 = 1.5*np.random.randn(300) + 5
y2 = 1.5*np.random.randn(300) + 3

x = np.append(x1,x2)
y = np.append(y1,y2)

# create fig and ax objects
fig, ax = plt.subplots()

# create the scatter plot with ax.scatter and customize
ax.scatter(x,y, alpha=0.3)  # x values, y values, alpha corresponds to the oppacity of the dots
ax.set_title('A scatter plot of two semi-focused arrays of random numbers')

# show the plot
plt.show()
```
