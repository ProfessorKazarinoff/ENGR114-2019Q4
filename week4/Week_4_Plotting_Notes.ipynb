### Create a first line plot

```python
# create a simple numpy array to plot
y = np.array([3,4,7,9,11])

# create the figure object, fig and the axes object, ax

fig, ax = plt.subplots()

# create the plot with ax.plot()
ax.plot(y)

# show the plot
plt.show()
```

### Build a line plot with a couple customizations
 
  * title
  * x-axis label
  * y-axis label
  * red line

```python
# Create the data to plot
y = np.array([1,3,4,7,8])
x = np.array([2,4,6,8,10])

# Create the fig, ax objects
fig, ax = plt.subplots()

# Build the plot and add the customizations
ax.plot(x,y,'r')
ax.set_title('A Line Plot with a Red Line')
ax.set_xlabel('x-axis label')
ax.set_ylabel('y-axis label')

# Show the plot
plt.show()
```

### Create a line plot with two lines, a legend and save it

 * sine and cosine functions, sine red and cos blue
 * x-axis label and y-axis label
 * title
 * legend
 * save the plot

```python
# Create the data
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the fig, ax objects
fig, ax = plt.subplots()

# Build and customize the plot
ax.plot(x,y1,'r')
ax.plot(x,y2,'b')
ax.set_xlabel('Angles 0 to 4pi')
ax.set_ylabel('Sine and Cosine')
ax.set_title('Line plot with two lines and a legend')
ax.legend(['sine','cosine'])

# Save and Show the plot
plt.savefig('plot_with_two_lines.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Build a multi-line plot and specify tick locations and tick labels

```python
# Create the data
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the fig, ax objects
fig, ax = plt.subplots()

# Build and customize the plot
ax.plot(x,y1,'r')
ax.plot(x,y2,'b')
ax.set_xlabel('Angles 0 to 4pi')
ax.set_ylabel('Sine and Cosine')
ax.set_title('Line plot with two lines and a legend')
ax.legend(['sine','cosine'])
ax.set_xticks(np.arange(0,4*np.pi+np.pi/2,np.pi/2))
ax.set_xticklabels(['0','$\pi/2$','pi','$3\pi/2$','2pi','5pi/2','3pi','7pi/2','4pi'])

# Save and Show the plot
plt.savefig('plot_with_two_lines.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Create a plot of sin, cos, tan and limit the range of the axes

```python
# Create the data
x = np.arange(0,np.pi*4,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create the fig, ax objects
fig, ax = plt.subplots()

# Build and customize the plot
ax.plot(x,y1,'r')
ax.plot(x,y2,'b')
ax.plot(x,y3,'y')
ax.set_xlim([0, np.pi*2])
ax.set_ylim([-3, 3])
ax.set_xlabel('Angles 0 to 4pi')
ax.set_ylabel('Sine and Cosine')
ax.set_title('Line plot with three lines and a legend')
ax.legend(['sine','cosine','tangent'], loc='upper right')

# Save and Show the plot
plt.savefig('plot_with_three_lines.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Build a pie chart with ax.pie

```python
# Create Data
labels = ['Windows','MacOS','Linux']
sizes = [13, 4, 2]
explode = [0.01, 0.01, 0.5]

# Create the fig, ax
fig, ax = plt.subplots()

# Build the pie chart with ax.pie and add customizations
ax.pie(sizes, labels=labels, autopct='%1.0f%%', explode=explode)
ax.axis('equal')
ax.set_title('Pie Chart with one piece exploded out')

# Save and show the plot
plt.savefig('pie_chart.png', dpi=73, bbox_inches='tight')
```

### Create bar chart with ax.bar

```python
# Create data
heights=[3, 11, 6]
bar_locs = [0,1,2]
bar_labels = ['less than 20min','between 20 and 40 min','greater than 20min']

# fig, ax objects
fig, ax = plt.subplots()

# ax.bar and customize the bar chart
ax.bar(bar_locs,heights, align='center')
ax.set_title('Student commute time ranges for ENGR114')
ax.set_xticks(bar_locs)
ax.set_xticklabels(bar_labels)

# save and show the plot
plt.savefig('bar_plot.jpg',dpi=72,bbox_inches='tight')
plt.show()
```

### Create a histogram with ax.hist

```python
# Create array of numbers that have a normal distribution
mu = 20
sigma = 5
x = np.random.normal(mu, sigma, size=500)

# fig, ax objects
fig, ax = plt.subplots()

# Create the histogram with ax.hist and customize
ax.hist(x, 10)    # 10 bins or bars on the histogram
ax.set_xlabel('bin range')
ax.set_ylabel('frequency')
ax.set_title('Histogram of a random normal distribution')

# Save and show the plot
plt.savefig('histogram.png', dpi=72, bbox_inches='tight')
plt.show()
```

### Create a scatter plot with ax.scatter

```python
# Create arrays of points to plot
n = 300
x1 = 2*np.random.randn(n) + 5
y1 = 2*np.random.randn(n) + 3

x2 = 4*np.random.randn(n) + 20
y2 = 4*np.random.randn(n) + 12

x = np.append(x1,x2)
y = np.append(y1,y2)

# fig, ax
fig, ax = plt.subplots()

# ax.scatter to plot the points and custumize
ax.scatter(x,y, alpha=0.1)
ax.set_title('A scatter plot of two semi-focused array of random numbers')

# Save and show the plot
plt.savefig('scatterplot.png',dpi=72,bbox_inches='tight')
plt.show()
```
